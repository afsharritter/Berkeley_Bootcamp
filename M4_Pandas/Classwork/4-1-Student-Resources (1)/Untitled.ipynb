```python
import pandas as pd
```

```python
raw_data_info = {
    "customer_id": [112, 403, 999, 543, 123],
    "name": ["bob", "john", "luke", 'arlene', 'jesus'],
    "email": ["bob@gmail.com", "john@gmail.com", "luke@gmail.com", 'arlene@gmail.com', 'jesus@gmail.com'],
}
info_df = pd.DataFrame(raw_data_info, columns=["customer_id", "name", "email"])
info_df
```

```python
raw_data_items = {
    "customer_id": [112, 403, 999, 543, 123],
    "item": ["chips", "soda", "smokes", 'H2O', 'crackers'],
    "cost": [2.5, 2, 11, 3, 1]
}
items_df = pd.DataFrame(raw_data_items, columns=["customer_id", "item", "cost"])
items_df
```

```python
merge_df = pd.merge(info_df, items_df, on="customer_id", how="right")
merge_df
```
