```python
sales_df = pd.DataFrame({"item": ["tomatoes", "onions", "potatoes", "tomatoes", "mushrooms"],
                "quantity": [15, 26, 10, 12, 2],
                "total_price": [50.25, 39.52, 50.01, 40.20, 8.46]})
sales_df.head()
```

```python
sales_df["quantity"] > 10
```

```python
filter = sales_df["quantity"] > 10
sales_df.loc[filter]
```

```python
sales_df.loc[sales_df["quantity"] > 10]
```

```python
sales_df.loc[
    (sales_df["quantity"] > 10) &
    (sales_df["total_price"] < 45)
     ]
```

```python
sales_df["unit_price"] = sales_df["total_price"]/sales_df["quantity"]
sales_df
```

```python
sales_df.loc[
    (sales_df["quantity"] > 10) |
    (sales_df["total_price"] < 45)
     ]
```

```python
min_total_price = sales_df["total_price"].min()
min_total_price_row = sales_df.loc[sales_df["total_price"] == min_total_price]
min_total_price_row
```

```python
min_unit_price = sales_df["unit_price"].min()
min_unit_price_row = sales_df.loc[sales_df["unit_price"] == min_unit_price]
min_unit_price_row
```

```python
sales_df.loc[sales_df["item"] == "tomatoes", ["total_price"]].mean()
```
